```python
df_2022 = df_2022[(df_2022['Code departement']==75) | (df_2022['Code departement']==92)| (df_2022['Code departement']==93) | (df_2022['Code departement']==94) | (df_2022['Code departement']==77) | (df_2022['Code departement']==78) | (df_2022['Code departement']==91)| (df_2022['Code departement']==95)]
df_2022 = df_2022[(df_2022['Nature mutation'] =='Vente') &(df_2022['Type local']!='Local industriel. commercial ou assimilé') & (df_2022['Type local']!='Dépendance') & (df_2022['Type local'].notna())]
df_2022 = df_2022[df_2022['Valeur fonciere'].notna()]
#df_2022 = df_2022.drop_duplicates()
df_2022
```

```python
df_2021 = df_2021[(df_2021['Code departement']==75) | (df_2021['Code departement']==92)| (df_2021['Code departement']==93) | (df_2021['Code departement']==94) | (df_2021['Code departement']==77) | (df_2021['Code departement']==78) | (df_2021['Code departement']==91)| (df_2021['Code departement']==95)]
df_2021 = df_2021[(df_2021['Nature mutation'] =='Vente') &(df_2021['Type local']!='Local industriel. commercial ou assimilé') & (df_2021['Type local']!='Dépendance') & (df_2021['Type local'].notna())]
df_2021 = df_2021[df_2021['Valeur fonciere'].notna()]
#df_2021 = df_2021.drop_duplicates()
df_2021
```

```python
df_2020 = df_2020[(df_2020['Code departement']==75) | (df_2020['Code departement']==92)| (df_2020['Code departement']==93) | (df_2020['Code departement']==94) | (df_2020['Code departement']==77) | (df_2020['Code departement']==78) | (df_2020['Code departement']==91)| (df_2020['Code departement']==95)]
df_2020 = df_2020[(df_2020['Nature mutation'] =='Vente') &(df_2020['Type local']!='Local industriel. commercial ou assimilé') & (df_2020['Type local']!='Dépendance') & (df_2020['Type local'].notna())]
df_2020 = df_2020[df_2020['Valeur fonciere'].notna()]
#df_2020 = df_2020.drop_duplicates()
df_2020
```

```python
df_total = pd.concat([df_2022, df_2021, df_2020], axis=0)
df_total = df_total.reset_index()
df_total
```

```python
df_total['Valeur fonciere'] = df_total['Valeur fonciere'].str.replace(',','.')
df_total['Valeur fonciere'] = df_total['Valeur fonciere'].astype(float)
df_total['Surface reelle bati'] = df_total['Surface reelle bati'].astype(float)
df_total
```

```python
df_total_group = df_total.groupby(['Valeur fonciere', 'Commune','Code departement',	'Code commune', 'Type local', 'Date mutation']).agg({'Surface reelle bati': [np.count_nonzero, np.max], 'Nombre pieces principales':'max'})
df_total_group.reset_index(inplace=True)
#df_total_group['Surface reelle bati'].value_counts()#sort_values(by='Surface reelle bati', ascending=False)
df_total_group.columns= ['_'.join(col) for col in df_total_group.columns.values]
df_total_group = df_total_group[df_total_group['Surface reelle bati_count_nonzero']==1] #excluded count'Surface realle bati'>1 which means it's a building
df_total_group
```

```python
df_total_group = df_total_group.rename(columns={'Valeur fonciere_':'Valeur fonciere', 'Commune_':'Commune','Code departement_':'Code departement', 'Code commune_':'Code commune', 'Type local_':'Type local', 'Date mutation_':'Date mutation', 
                               'Surface reelle bati_amax':'Surface reelle bati', 'Nombre pieces principales_max':'Nombre pieces principales'})
df_total_group = df_total_group[['Date mutation','Valeur fonciere', 'Commune', 'Code departement', 'Code commune', 'Type local', 'Surface reelle bati','Nombre pieces principales']]
df_total_group
```

```python
df_total_group['Valeur fonciere'] = df_total_group['Valeur fonciere'].astype(float)
df_total_group['Surface reelle bati'] = df_total_group['Surface reelle bati'].astype(float)
df_total_group['prix m2'] = df_total_group['Valeur fonciere']/df_total_group['Surface reelle bati']
df_total_group.columns
```

```python
df_final = df_total_group[['Date mutation', 'Valeur fonciere', 'Surface reelle bati', 'prix m2',
       'Nombre pieces principales', 'Type local', 'Commune', 'Code commune',
       'Code departement']]
df_final
```

```python
df_final = df_final[(df_final['prix m2']>1000) &(df_final['prix m2']<20000)]
```

```python
df_final['Year'] = df_final['Date mutation'].apply(lambda x :x[6:10])
df_final.reset_index(inplace=True)
df_final.drop(['index'], axis = 1, inplace = True) 
df_final
```

```python
df_final_22 = df_final[df_final['Year']=='2022']
df_final_22.reset_index(inplace=True)
df_final_22.drop(['index'], axis = 1, inplace = True) 
df_final_22
```
